```python
# Stage 2 : Mass Balance and Flagging 
df["Difference"] = df["Quantity_in"] = df["Quantity_out"]

thresholds = {"kg": 1.0, "g": 20.0, "Ci": 2.0, "mCi": 100.0}
df["Flag"] = df.apply(
    lambda r: "CHECK" if abs (r["Difference"]) > thresholds.get(r["Unit"], 1) else "OK", 
    axis=1
)

print("\nFacilities with CHECK flag:")
print(df[df["Flag"] == "CHECK"].head())
```

```python
# Stage 3 : Vasualization 
# Goal:
# Visualize total radiation activity across different Saudi cities to identify which regions require stronger monitoring.

# Code:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset — synthetic data representing radiation activity per city
data = {
    'City': ['Makkah', 'Riyadh', 'Jeddah', 'Dammam', 'Madinah', 'Abha', 'Tabuk'],
    'Total_Radiation_Activity': [950, 870, 650, 520, 410, 280, 150]
}
df_city = pd.DataFrame(data)

# Sort cities from highest to lowest activity
df_city = df_city.sort_values(by='Total_Radiation_Activity', ascending=False)

# Create visualization
plt.figure(figsize=(9,6))
sns.barplot(x='City', y='Total_Radiation_Activity', data=df_city, palette='plasma')
plt.title("Radiation Activity by City (Highest to Lowest)")
plt.xlabel("City")
plt.ylabel("Total Radiation Activity")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# Explanation:
# This bar chart shows the total radiation activity in each city.
# Makkah has the highest activity level, followed by Riyadh and Jeddah.
# The gradual decrease from left to right indicates reduced radiation activity across regions.

# Solutions:
# - Prioritize safety inspections in Makkah, Riyadh, and Jeddah.
# - Investigate the reasons behind lower activity levels in Abha and Tabuk (fewer facilities or shipments).
# - Use this ranking to allocate inspection resources efficiently.
```

```python
# Goal:
# Show how radiation activity changes over time for each city.
# This helps detect which cities have consistent increases or sudden spikes in radiation levels.

# Code:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create synthetic time-series data
data = {
    'Date': pd.date_range(start='2025-01-01', periods=6, freq='M'),
    'Makkah': [850, 920, 950, 910, 880, 940],
    'Riyadh': [750, 810, 870, 880, 860, 870],
    'Jeddah': [600, 640, 650, 670, 680, 660],
}
df_trend = pd.DataFrame(data)

# Convert from wide to long format
df_long = df_trend.melt(id_vars='Date', var_name='City', value_name='Radiation')

# Plot line chart
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Radiation', hue='City', data=df_long, marker='o')
plt.title("Radiation Trend Over Time by City")
plt.xlabel("Date")
plt.ylabel("Radiation Activity Level")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='City')
plt.show()

# Explanation:
# Each line represents a city, showing how radiation levels changed monthly.
# Makkah shows consistent high activity; Riyadh is moderate; Jeddah is lower but stable.

# Solutions:
# - Investigate why Makkah’s activity remains high across all months.
# - Identify months with sharp increases to understand potential causes.
# - Maintain monitoring consistency for lower-activity cities like Jeddah.
```

```python
# Stage 4: Summary Report

# Goal:
# To generate a concise summary of the dataset, 
# highlighting the number of total and flagged (CHECK) records,
# and identifying the most frequently flagged materials.

total_records = len(df)
check_cases = len(df[df["Flag"] == "CHECK"])

print(f"Total records: {total_records}")
print(f"CHECK cases: {check_cases}")
print(f"Percentage of flagged cases: {check_cases/total_records*100:.2f}%")

print("\nTop materials with most CHECK flags:")
print(df[df["Flag"] == "CHECK"]["Material"].value_counts().head())

# Explanation:
# This step counts how many records exist in total, and how many were flagged as “CHECK”.
# Then, it identifies which materials were most often associated with anomalies.
# This helps prioritize which materials or facilities require deeper inspection.

# Insight:
# For example, if Uranium-235 appears most often in “CHECK” cases,
# it could mean either measurement inaccuracies or potential safety non-compliance.
```

```python
# Stage 5: Root Cause Analysis

# Goal:
# To calculate the average quantity difference for each flagged material,
# in order to identify which materials consistently show the largest imbalances.

problem_summary = (
    df[df["Flag"] == "CHECK"]
    .groupby("Material")["Difference"]
    .mean()
    .sort_values(ascending=False)
)
print("\nAverage difference per material (for CHECK cases):")
print(problem_summary)

# Explanation:
# This groups all flagged cases by material and computes the mean difference.
# Materials with higher average differences may indicate systematic errors
# in handling, logging, or radiation decay miscalculations.

# Insight:
# A pattern of high imbalance in the same material (e.g., Cesium-137)
# may signal the need for recalibration of measurement instruments
# or closer oversight of that material’s transport process.
```

```python
# Stage 7: Early Warning System (Anomaly Detection)

# Objective:
# Automatically detect sudden abnormal differences in nuclear material quantities
# and raise alerts for possible data errors or unreported activities.

# Step 1: Calculate mean and standard deviation for each material
stats = df.groupby("Material")["Difference"].agg(["mean", "std"]).reset_index()

# Step 2: Merge stats back to the main dataframe
df = df.merge(stats, on="Material", how="left")

# Step 3: Define anomaly threshold (± 2 * standard deviation)
df["Anomaly"] = abs(df["Difference"] - df["mean"]) > (2 * df["std"])

# Step 4: Show materials with anomalies
anomalies = df[df["Anomaly"] == True]
print("⚠️ Detected Anomalies:")
print(anomalies[["FacilityName", "Material", "Difference", "City"]])

# Optional: Visualize anomaly distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.scatter(df["Material"], df["Difference"], c=df["Anomaly"], cmap="coolwarm")
plt.title("Detected Anomalies in Material Quantities")
plt.xlabel("Material")
plt.ylabel("Difference (In - Out)")
plt.xticks(rotation=45)
plt.show()

# Explanation:
# - The system calculates the historical average difference per material.
# - If any new record exceeds ±2 standard deviations, it's marked as an anomaly.
# - This helps identify suspicious or inconsistent activity early.

# Recommendations:
# 1. Review flagged facilities for data accuracy.
# 2. Implement automatic alerts for these anomalies in real-time dashboards.
# 3. Train staff to report significant deviations immediately.
```
